##                                                    Data Visualization

### Data visualization is defined as a graphical representation that contains the information and the data.

### By using visual elements like charts, graphs, and maps, data visualization techniques provide an accessible way to see and understand trends, outliers, and patterns in data.

## Different Types of Analysis for Data Visualization

### Univariate Analysis: In the univariate analysis, we will be using a single feature to analyze almost all of its properties.

### Bivariate Analysis: When we compare the data between exactly 2 features then it is known as bivariate analysis.

### Multivariate Analysis: In the multivariate analysis, we will be comparing more than 2 variables.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# initializing the data

x = [10, 20, 30, 40]
y = [30, 22, 45, 50]
 
# plotting the data

plt.plot(x, y)
 
plt.show()
```

```python
# initializing the data

x = [10, 20, 30, 40]
y = [30, 22, 45, 50]
 
# plotting the data

plt.plot(x, y)
 
    
# Adding title to the plot

plt.title("Linear graph", fontsize=25, color="purple")

# Adding label on the y-axis

plt.ylabel('Y-Axis')
 
# Adding label on the x-axis

plt.xlabel('X-Axis')

# setting the labels of x-axis

plt.xticks(x, labels=["one", "two", "three", "four"])
 
# Adding legends

plt.legend(["Linear Graph"])

plt.show()
```

```python
x = [10, 20, 30, 40]
y = [30, 22, 45, 50]
 
# plotting the data


# Creating a new figure with width = 7 inches
# and height = 5 inches with face color as
# green, edgecolor as red and the line width
# of the edge as 7

fig = plt.figure(figsize =(7, 5), facecolor='g',
                 edgecolor='b', linewidth=7)

# Creating a new axes for the figure

a1 = fig.add_axes([1, 1, 1, 1])
a1.plot(x, y)
plt.show()
```

## Subplots

## Lineplot 
### is used to get understanding of distribution of 2 numerical variables, time series data

```python
d1.head(2)
```

```python
sns.lineplot(x="timepoint", y="signal",data=d1)
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal",data=d1, hue="region", style="event", markers=True)
#plt.legend(["lineplot"])  
plt.show()
```

```python
# in seaborn , the hue parameter denotes which column decides the kind of color, i.e. it's red, or green
```

## Barplot
### is used to understand about categorical variable

```python
sns.set(style="whitegrid")
sns.barplot(x="event", y="signal",data=d1, palette="rocket")
plt.show()
```

## Scatter plot
### is used on continous numerical values

```python
df1 = pd.read_csv('iris.csv')
```

```python
df1.head()
```

```python
sns.scatterplot(x = "SepalLengthCm", y ="PetalWidthCm", data =df1, color='green')

plt.xlabel('Sepal length')

plt.ylabel('Petal width')

plt.title('Scatter plot on Iris dataset')

plt.show()
```

## Pair plot
### Lets say we have n number of features in a data, Pair plot will create us a (n x n)

```python
sns.set_style("whitegrid")
sns.pairplot(df1, hue="Species", height=2)
plt.show()
```

## Heatmap

### Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix.

### Heatmap is also defined by the name of the shading matrix. Heatmaps in Seaborn can be plotted by using the seaborn.heatmap() function.

```python
df=pd.DataFrame(np.random.random((7,7)),columns=['a','b','c','d','e','f','g'])
sns.heatmap(df)
```

```python
sns.heatmap(df,annot=True,annot_kws={'size':7})
```
